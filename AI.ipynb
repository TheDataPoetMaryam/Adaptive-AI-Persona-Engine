# 🤖 Synthetic Persona Intelligence  
### Context-Aware Conversational AI System using Large Language Models

##### This project demonstrates the design of an adaptive AI conversation system capable of generating personality-driven responses using Large Language Models.

##### The system combines:
##### - Prompt Engineering
##### - Persona Configuration
##### - Conversation Memory
##### - API-based LLM Integration
##### - Context-aware Response Generation

```python
import requests
import json
import time
import os
```

```python
import os
from getpass import getpass


os.environ["OPENROUTER_API_KEY"] = getpass(
    "Enter OpenRouter API Key: "
)

API_KEY = os.getenv("OPENROUTER_API_KEY")
API_URL = "https://openrouter.ai/api/v1/chat/completions"
```

```python
API_KEY = os.getenv("OPENROUTER_API_KEY")                                                                                  
MODEL = "meta-llama/llama-3.3-70b-instruct:free"
```

```python
personas = {

    "empathetic":
    "You are supportive, thoughtful and emotionally aware.",

    "analytical":
    "You provide logical, structured and concise responses.",

    "creative":
    "You respond with imagination and expressive storytelling."

}


persona_choice = input(
    "Choose persona (empathetic / analytical / creative): "
)


system_msg = personas.get(
    persona_choice,
    "You are a helpful AI assistant."
)
```

## 🧠 System Workflow

##### User Input  
##### ↓  
##### Persona Selection  
##### ↓  
##### Prompt Configuration  
##### ↓  
##### Conversation Memory  
##### ↓  
##### LLM Processing  
##### ↓  
##### Context-Aware Response

```python
scenario = input("Enter conversation scenario/context: ")

try:
    with open("memory.json", "r") as file:
        conversation = json.load(file)

except FileNotFoundError:
    conversation = []
```

```python
persona_prompt = f"""

You are an adaptive AI assistant.

Persona Behavior:
{system_msg}

Context:
{scenario}

Rules:
- Maintain personality consistency
- Remember conversation context
- Generate natural responses
- Stay aligned with assigned behavior

"""
```

```python
# Initialize conversation with system instructions

conversation = [
    {
        "role": "system",
        "content": persona_prompt
    }
]

print("✨ AI Persona System Started")
print("Type 'exit' to stop\n")


while True:
    user_input = input("You: ")
    if user_input.lower() == "exit":
        print("Conversation ended.")
        break

    # Add user message
    conversation.append(
        {
            "role": "user",
            "content": user_input
        }
    )

    response = requests.post(
        url=API_URL,
        headers={
            "Authorization": f"Bearer {API_KEY}",
            "Content-Type": "application/json"
        },

        json={
            "model": MODEL,
            "messages": conversation
        }
    )

    data = response.json()
    if "choices" in data:
        reply = data["choices"][0]["message"]["content"]
        print("AI is thinking...", end="\r")
        time.sleep(1)
        print("\nAI:", reply)
        
        # Store AI response
        conversation.append(
            {
                "role": "assistant",
                "content": reply
            }
        )

        # Save memory
        with open("memory.json", "w") as file:
            json.dump(
                conversation,
                file,
                indent=4
            )

    else:

        error_message = data.get("error", {}).get(
            "message",
            "Unknown error occurred"
        )

        print("⚠️ API Response Error:")
        print(json.dumps(data, indent=4))
```
